```
#Câu 1

def LuuFile(path, data):
    file = open(path, 'a', encoding='utf-8')
    file.writelines(data)
    file.writelines("\n")
    file.close()

def DocFile(path):
    arrProduct = []
    file = open(path, 'r', encoding='utf-8')
    for line in file:
        data = line.strip()
        arr = data.split(';')
        arrProduct.append(arr)
    file.close()
    return arrProduct


# --- Nhập sản phẩm ---
masp = input("Nhập mã SP: ")
tensp = input("Nhập tên SP: ")
dongia = float(input("Nhập giá: "))

line = masp + ";" + tensp + ";" + str(dongia)
LuuFile("database.txt", line)

# --- Đọc và xuất sản phẩm ---
dssp = DocFile("database.txt")

def XuatSanPham(dssp):
    print("Mã SP\tTên SP\tĐơn Giá")
    for row in dssp:
        for element in row:
            print(element, end='\t')
        print()
    print()

print("=== Danh sách sản phẩm ban đầu ===")
XuatSanPham(dssp)

# --- Sắp xếp giảm dần theo đơn giá ---
def SortSp(dssp):
    for i in range(len(dssp)):
        for j in range(i + 1, len(dssp)):
            if float(dssp[i][2]) < float(dssp[j][2]):
                dssp[i], dssp[j] = dssp[j], dssp[i]

SortSp(dssp)
print("=== Danh sách sản phẩm sau khi sắp xếp giảm giá ===")
XuatSanPham(dssp)
```

```
#Câu 3

# --- Bước 1: Tạo file employees.xml ---
xml_data = """<?xml version="1.0" encoding="UTF-8"?>
<employees>
    <employee>
        <id>1</id>
        <name>Trần Duy Thanh</name>
    </employee>
    <employee>
        <id>2</id>
        <name>Lê Hoành Sử</name>
    </employee>
    <employee>
        <id>3</id>
        <name>Hồ Trung Thành</name>
    </employee>
</employees>
"""

# Ghi nội dung XML vào file
with open("employees.xml", "w", encoding="utf-8") as f:
    f.write(xml_data)

print("✅ Đã tạo file employees.xml thành công!\n")

# --- Bước 2: Đọc và hiển thị dữ liệu từ XML ---
from xml.dom.minidom import parse
import xml.dom.minidom

# Mở file XML bằng DOM parser
DOMTree = xml.dom.minidom.parse("employees.xml")
collection = DOMTree.documentElement

# Lấy tất cả các thẻ <employee>
employees = collection.getElementsByTagName("employee")

print("=== DANH SÁCH NHÂN VIÊN TRONG FILE XML ===")
for employee in employees:
    tag_id = employee.getElementsByTagName("id")[0]
    id = tag_id.childNodes[0].data

    tag_name = employee.getElementsByTagName("name")[0]
    name = tag_name.childNodes[0].data

    print(f"ID: {id}\tTên: {name}")
```

```

#Câu 9
import os

FILE_NAME = "sanpham.txt"

class SanPham:
    def __init__(self, ma, ten, dongia, madm):
        self.ma = ma
        self.ten = ten
        self.dongia = dongia
        self.madm = madm

    def __str__(self):
        return f"{self.ma},{self.ten},{self.dongia},{self.madm}"

def them_san_pham():
    ma = input("Mã sản phẩm: ")
    ten = input("Tên sản phẩm: ")
    dongia = float(input("Đơn giá: "))
    madm = input("Mã danh mục: ")
    sp = SanPham(ma, ten, dongia, madm)
    with open(FILE_NAME, "a", encoding="utf-8") as f:
        f.write(str(sp) + "\n")

def doc_file():
    ds = []
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r", encoding="utf-8") as f:
            for line in f:
                parts = line.strip().split(",")
                if len(parts) == 4:
                    ds.append(SanPham(*parts))
    return ds

def hien_thi():
    ds = doc_file()
    print("Mã\tTên\tĐơn giá\tMã DM")
    for sp in ds:
        print(f"{sp.ma}\t{sp.ten}\t{sp.dongia}\t{sp.madm}")

def tim_kiem():
    ten = input("Nhập tên cần tìm: ").lower()
    ds = doc_file()
    kq = [sp for sp in ds if ten in sp.ten.lower()]
    for sp in kq:
        print(sp.ma, sp.ten, sp.dongia, sp.madm)

def sap_xep():
    ds = doc_file()
    ds.sort(key=lambda x: float(x.dongia))
    for sp in ds:
        print(sp)

# Giao diện menu
while True:
    print("\n--- QUẢN LÝ SẢN PHẨM ---")
    print("1. Thêm sản phẩm")
    print("2. Xem danh sách")
    print("3. Tìm kiếm")
    print("4. Sắp xếp theo giá")
    print("0. Thoát")
    chon = input("Chọn: ")
    if chon == "1": them_san_pham()
    elif chon == "2": hien_thi()
    elif chon == "3": tim_kiem()
    elif chon == "4": sap_xep()
    elif chon == "0": break
```

```
#Câu 10
import json, os

FILE = "sinhvien.json"

def doc_file():
    if not os.path.exists(FILE):
        return []
    with open(FILE, "r", encoding="utf-8") as f:
        return json.load(f)

def ghi_file(ds):
    with open(FILE, "w", encoding="utf-8") as f:
        json.dump(ds, f, ensure_ascii=False, indent=4)

def them_sv():
    ds = doc_file()
    ma = input("Mã SV: ")
    ten = input("Tên: ")
    ns = int(input("Năm sinh: "))
    malop = input("Mã lớp: ")
    ds.append({"MaSV": ma, "Ten": ten, "NamSinh": ns, "MaLop": malop})
    ghi_file(ds)

def hien_thi():
    ds = doc_file()
    for sv in ds:
        print(sv)

def sap_xep():
    ds = doc_file()
    ds.sort(key=lambda x: x["NamSinh"])
    for sv in ds:
        print(sv)

# Menu
while True:
    print("\n--- QUẢN LÝ SINH VIÊN ---")
    print("1. Thêm sinh viên")
    print("2. Xem danh sách")
    print("3. Sắp xếp theo năm sinh")
    print("0. Thoát")
    chon = input("Chọn: ")
    if chon == "1": them_sv()
    elif chon == "2": hien_thi()
    elif chon == "3": sap_xep()
    elif chon == "0": break
```

```
#Câu 11
from openpyxl import Workbook, load_workbook
import os

FILE = "nhanvien.xlsx"

def luu_excel():
    wb = Workbook()
    ws = wb.active
    ws.append(["Mã", "Tên", "Tuổi"])
    while True:
        ma = input("Mã NV: ")
        ten = input("Tên: ")
        tuoi = int(input("Tuổi: "))
        ws.append([ma, ten, tuoi])
        tiep = input("Nhập tiếp? (y/n): ")
        if tiep.lower() != 'y':
            break
    wb.save(FILE)
    print("Đã lưu Excel thành công!")

def doc_excel():
    if not os.path.exists(FILE):
        print("Chưa có file Excel!")
        return
    wb = load_workbook(FILE)
    ws = wb.active
    for row in ws.iter_rows(values_only=True):
        print(row)

def sap_xep_excel():
    wb = load_workbook(FILE)
    ws = wb.active
    data = list(ws.iter_rows(values_only=True))
    header, nv = data[0], data[1:]
    nv.sort(key=lambda x: x[2])
    wb2 = Workbook()
    ws2 = wb2.active
    ws2.append(header)
    for r in nv:
        ws2.append(r)
    wb2.save("nhanvien_sapxep.xlsx")
    print("Đã lưu file nhanvien_sapxep.xlsx!")

# Menu
while True:
    print("\n--- QUẢN LÝ NHÂN VIÊN EXCEL ---")
    print("1. Nhập & lưu Excel")
    print("2. Đọc danh sách")
    print("3. Sắp xếp theo tuổi")
    print("0. Thoát")
    chon = input("Chọn: ")
    if chon == "1": luu_excel()
    elif chon == "2": doc_excel()
    elif chon == "3": sap_xep_excel()
    elif chon == "0": break
```

```
#Câu 12
import csv
import random

# --- BƯỚC 1: Tạo file CSV ---
def tao_file_csv(ten_file):
    with open(ten_file, 'w', newline='') as f:
        writer = csv.writer(f, delimiter=';')
        for _ in range(10):  # 10 dòng
            dong = [random.randint(0, 100) for _ in range(10)]  # 10 số ngẫu nhiên
            writer.writerow(dong)
    print(f"Đã tạo file {ten_file} thành công!")

# --- BƯỚC 2: Đọc file và tính tổng mỗi dòng ---
def doc_va_tinh_tong(ten_file):
    with open(ten_file, 'r') as f:
        reader = csv.reader(f, delimiter=';')
        print("Tổng giá trị mỗi dòng:")
        for i, dong in enumerate(reader, start=1):
            # Chuyển từng phần tử sang int, bỏ qua ô trống nếu có
            tong = sum(int(x) for x in dong if x.strip() != '')
            print(f"Dòng {i}: {tong}")

# --- CHƯƠNG TRÌNH CHÍNH ---
ten_file = "dulieu.csv"
tao_file_csv(ten_file)
doc_va_tinh_tong(ten_file)
```

```
#Câu 13
import xml.etree.ElementTree as ET
from collections import Counter
import os

# --- Tạo file XML nhóm thiết bị ---
def tao_file_nhom_thiet_bi(file_nhom):
    xml_nhom_data = """<?xml version="1.0" encoding="UTF-8"?>
<nhomthietbis>
    <nhom>
        <ma>n1</ma>
        <ten>Điện thoại</ten>
    </nhom>
    <nhom>
        <ma>n2</ma>
        <ten>Máy tính bảng</ten>
    </nhom>
    <nhom>
        <ma>n3</ma>
        <ten>Laptop</ten>
    </nhom>
</nhomthietbis>
"""
    with open(file_nhom, "w", encoding="utf-8") as f:
        f.write(xml_nhom_data)
    print(f"✅ Đã tạo file {file_nhom} thành công!")

# --- Tạo file XML thiết bị ---
def tao_file_thiet_bi(file_tb):
    xml_tb_data = """<?xml version="1.0" encoding="UTF-8"?>
<thietbis>
    <thietbi manhom="n1">
        <ma>dt01</ma>
        <ten>iPhone 13</ten>
    </thietbi>
    <thietbi manhom="n1">
        <ma>dt02</ma>
        <ten>Samsung S22</ten>
    </thietbi>
    <thietbi manhom="n2">
        <ma>mtb01</ma>
        <ten>iPad Air</ten>
    </thietbi>
    <thietbi manhom="n3">
        <ma>lt01</ma>
        <ten>Macbook Pro</ten>
    </thietbi>
    <thietbi manhom="n3">
        <ma>lt02</ma>
        <ten>Dell XPS 15</ten>
    </thietbi>
</thietbis>
"""
    with open(file_tb, "w", encoding="utf-8") as f:
        f.write(xml_tb_data)
    print(f"✅ Đã tạo file {file_tb} thành công!")


# --- Đọc file XML nhóm thiết bị ---
def doc_nhom_thiet_bi(file_nhom):
    tree = ET.parse(file_nhom)
    root = tree.getroot()
    nhoms = {}
    for nhom in root.findall('nhom'):
        ma = nhom.find('ma').text
        ten = nhom.find('ten').text
        nhoms[ma] = ten
    return nhoms

# --- Đọc file XML thiết bị ---
def doc_thiet_bi(file_tb):
    tree = ET.parse(file_tb)
    root = tree.getroot()
    ds_tb = []
    for tb in root.findall('thietbi'):
        manhom = tb.attrib['manhom']
        ma = tb.find('ma').text
        ten = tb.find('ten').text
        ds_tb.append({'manhom': manhom, 'ma': ma, 'ten': ten})
    return ds_tb

# --- Hiển thị danh sách nhóm ---
def hien_thi_nhom(nhoms):
    print("== Danh sách nhóm thiết bị ==")
    for ma, ten in nhoms.items():
        print(f"{ma} - {ten}")
    print()

# --- Hiển thị toàn bộ thiết bị ---
def hien_thi_thiet_bi(ds_tb):
    print("== Danh sách thiết bị ==")
    for tb in ds_tb:
        print(f"Mã: {tb['ma']}, Tên: {tb['ten']}, Nhóm: {tb['manhom']}")
    print()

# --- Lọc thiết bị theo mã nhóm ---
def loc_theo_nhom(ds_tb, manhom):
    print(f"== Thiết bị thuộc nhóm {manhom} ==")
    found = False
    for tb in ds_tb:
        if tb['manhom'] == manhom:
            print(f"  {tb['ma']} - {tb['ten']}")
            found = True
    if not found:
        print(f"  Không tìm thấy thiết bị nào thuộc nhóm {manhom}")
    print()


# --- Tìm nhóm có nhiều thiết bị nhất ---
def nhom_nhieu_thiet_bi(ds_tb):
    if not ds_tb:
        print("== Không có dữ liệu thiết bị để tìm nhóm nhiều nhất ==")
        return

    dem = Counter([tb['manhom'] for tb in ds_tb])
    # Handle case where Counter is empty
    if not dem:
        print("== Không có dữ liệu thiết bị để tìm nhóm nhiều nhất ==")
        return

    nhom_max = dem.most_common(1)[0]
    print(f"== Nhóm có nhiều thiết bị nhất ==")
    print(f"Nhóm: {nhom_max[0]} - Số lượng: {nhom_max[1]}")
    print()

# --- Chương trình chính ---
def main():
    file_nhom = "nhomthietbi.xml"
    file_tb = "thietbi.xml"

    # Tạo file nếu chưa tồn tại
    if not os.path.exists(file_nhom):
        tao_file_nhom_thiet_bi(file_nhom)
    if not os.path.exists(file_tb):
        tao_file_thiet_bi(file_tb)

    nhoms = doc_nhom_thiet_bi(file_nhom)
    ds_tb = doc_thiet_bi(file_tb)

    hien_thi_nhom(nhoms)
    hien_thi_thiet_bi(ds_tb)
    loc_theo_nhom(ds_tb, "n1")
    nhom_nhieu_thiet_bi(ds_tb)

if __name__ == "__main__":
    main()
```
