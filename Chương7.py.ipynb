```
#C√¢u 1

def LuuFile(path, data):
    file = open(path, 'a', encoding='utf-8')
    file.writelines(data)
    file.writelines("\n")
    file.close()

def DocFile(path):
    arrProduct = []
    file = open(path, 'r', encoding='utf-8')
    for line in file:
        data = line.strip()
        arr = data.split(';')
        arrProduct.append(arr)
    file.close()
    return arrProduct


# --- Nh·∫≠p s·∫£n ph·∫©m ---
masp = input("Nh·∫≠p m√£ SP: ")
tensp = input("Nh·∫≠p t√™n SP: ")
dongia = float(input("Nh·∫≠p gi√°: "))

line = masp + ";" + tensp + ";" + str(dongia)
LuuFile("database.txt", line)

# --- ƒê·ªçc v√† xu·∫•t s·∫£n ph·∫©m ---
dssp = DocFile("database.txt")

def XuatSanPham(dssp):
    print("M√£ SP\tT√™n SP\tƒê∆°n Gi√°")
    for row in dssp:
        for element in row:
            print(element, end='\t')
        print()
    print()

print("=== Danh s√°ch s·∫£n ph·∫©m ban ƒë·∫ßu ===")
XuatSanPham(dssp)

# --- S·∫Øp x·∫øp gi·∫£m d·∫ßn theo ƒë∆°n gi√° ---
def SortSp(dssp):
    for i in range(len(dssp)):
        for j in range(i + 1, len(dssp)):
            if float(dssp[i][2]) < float(dssp[j][2]):
                dssp[i], dssp[j] = dssp[j], dssp[i]

SortSp(dssp)
print("=== Danh s√°ch s·∫£n ph·∫©m sau khi s·∫Øp x·∫øp gi·∫£m gi√° ===")
XuatSanPham(dssp)
```

```

#C√¢u 9
import os

FILE_NAME = "sanpham.txt"

class SanPham:
    def __init__(self, ma, ten, dongia, madm):
        self.ma = ma
        self.ten = ten
        self.dongia = dongia
        self.madm = madm

    def __str__(self):
        return f"{self.ma},{self.ten},{self.dongia},{self.madm}"

def them_san_pham():
    ma = input("M√£ s·∫£n ph·∫©m: ")
    ten = input("T√™n s·∫£n ph·∫©m: ")
    dongia = float(input("ƒê∆°n gi√°: "))
    madm = input("M√£ danh m·ª•c: ")
    sp = SanPham(ma, ten, dongia, madm)
    with open(FILE_NAME, "a", encoding="utf-8") as f:
        f.write(str(sp) + "\n")

def doc_file():
    ds = []
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r", encoding="utf-8") as f:
            for line in f:
                parts = line.strip().split(",")
                if len(parts) == 4:
                    ds.append(SanPham(*parts))
    return ds

def hien_thi():
    ds = doc_file()
    print("M√£\tT√™n\tƒê∆°n gi√°\tM√£ DM")
    for sp in ds:
        print(f"{sp.ma}\t{sp.ten}\t{sp.dongia}\t{sp.madm}")

def tim_kiem():
    ten = input("Nh·∫≠p t√™n c·∫ßn t√¨m: ").lower()
    ds = doc_file()
    kq = [sp for sp in ds if ten in sp.ten.lower()]
    for sp in kq:
        print(sp.ma, sp.ten, sp.dongia, sp.madm)

def sap_xep():
    ds = doc_file()
    ds.sort(key=lambda x: float(x.dongia))
    for sp in ds:
        print(sp)

# Giao di·ªán menu
while True:
    print("\n--- QU·∫¢N L√ù S·∫¢N PH·∫®M ---")
    print("1. Th√™m s·∫£n ph·∫©m")
    print("2. Xem danh s√°ch")
    print("3. T√¨m ki·∫øm")
    print("4. S·∫Øp x·∫øp theo gi√°")
    print("0. Tho√°t")
    chon = input("Ch·ªçn: ")
    if chon == "1": them_san_pham()
    elif chon == "2": hien_thi()
    elif chon == "3": tim_kiem()
    elif chon == "4": sap_xep()
    elif chon == "0": break
```

```
#C√¢u 10
import json, os

FILE = "sinhvien.json"

def doc_file():
    if not os.path.exists(FILE):
        return []
    with open(FILE, "r", encoding="utf-8") as f:
        return json.load(f)

def ghi_file(ds):
    with open(FILE, "w", encoding="utf-8") as f:
        json.dump(ds, f, ensure_ascii=False, indent=4)

def them_sv():
    ds = doc_file()
    ma = input("M√£ SV: ")
    ten = input("T√™n: ")
    ns = int(input("NƒÉm sinh: "))
    malop = input("M√£ l·ªõp: ")
    ds.append({"MaSV": ma, "Ten": ten, "NamSinh": ns, "MaLop": malop})
    ghi_file(ds)

def hien_thi():
    ds = doc_file()
    for sv in ds:
        print(sv)

def sap_xep():
    ds = doc_file()
    ds.sort(key=lambda x: x["NamSinh"])
    for sv in ds:
        print(sv)

# Menu
while True:
    print("\n--- QU·∫¢N L√ù SINH VI√äN ---")
    print("1. Th√™m sinh vi√™n")
    print("2. Xem danh s√°ch")
    print("3. S·∫Øp x·∫øp theo nƒÉm sinh")
    print("0. Tho√°t")
    chon = input("Ch·ªçn: ")
    if chon == "1": them_sv()
    elif chon == "2": hien_thi()
    elif chon == "3": sap_xep()
    elif chon == "0": break
```

```
#C√¢u 11
from openpyxl import Workbook, load_workbook
import os

FILE = "nhanvien.xlsx"

def luu_excel():
    wb = Workbook()
    ws = wb.active
    ws.append(["M√£", "T√™n", "Tu·ªïi"])
    while True:
        ma = input("M√£ NV: ")
        ten = input("T√™n: ")
        tuoi = int(input("Tu·ªïi: "))
        ws.append([ma, ten, tuoi])
        tiep = input("Nh·∫≠p ti·∫øp? (y/n): ")
        if tiep.lower() != 'y':
            break
    wb.save(FILE)
    print("ƒê√£ l∆∞u Excel th√†nh c√¥ng!")

def doc_excel():
    if not os.path.exists(FILE):
        print("Ch∆∞a c√≥ file Excel!")
        return
    wb = load_workbook(FILE)
    ws = wb.active
    for row in ws.iter_rows(values_only=True):
        print(row)

def sap_xep_excel():
    wb = load_workbook(FILE)
    ws = wb.active
    data = list(ws.iter_rows(values_only=True))
    header, nv = data[0], data[1:]
    nv.sort(key=lambda x: x[2])
    wb2 = Workbook()
    ws2 = wb2.active
    ws2.append(header)
    for r in nv:
        ws2.append(r)
    wb2.save("nhanvien_sapxep.xlsx")
    print("ƒê√£ l∆∞u file nhanvien_sapxep.xlsx!")

# Menu
while True:
    print("\n--- QU·∫¢N L√ù NH√ÇN VI√äN EXCEL ---")
    print("1. Nh·∫≠p & l∆∞u Excel")
    print("2. ƒê·ªçc danh s√°ch")
    print("3. S·∫Øp x·∫øp theo tu·ªïi")
    print("0. Tho√°t")
    chon = input("Ch·ªçn: ")
    if chon == "1": luu_excel()
    elif chon == "2": doc_excel()
    elif chon == "3": sap_xep_excel()
    elif chon == "0": break
```

```
#C√¢u 12
import csv
import random

# --- B∆Ø·ªöC 1: T·∫°o file CSV ---
def tao_file_csv(ten_file):
    with open(ten_file, 'w', newline='') as f:
        writer = csv.writer(f, delimiter=';')
        for _ in range(10):  # 10 d√≤ng
            dong = [random.randint(0, 100) for _ in range(10)]  # 10 s·ªë ng·∫´u nhi√™n
            writer.writerow(dong)
    print(f"ƒê√£ t·∫°o file {ten_file} th√†nh c√¥ng!")

# --- B∆Ø·ªöC 2: ƒê·ªçc file v√† t√≠nh t·ªïng m·ªói d√≤ng ---
def doc_va_tinh_tong(ten_file):
    with open(ten_file, 'r') as f:
        reader = csv.reader(f, delimiter=';')
        print("T·ªïng gi√° tr·ªã m·ªói d√≤ng:")
        for i, dong in enumerate(reader, start=1):
            # Chuy·ªÉn t·ª´ng ph·∫ßn t·ª≠ sang int, b·ªè qua √¥ tr·ªëng n·∫øu c√≥
            tong = sum(int(x) for x in dong if x.strip() != '')
            print(f"D√≤ng {i}: {tong}")

# --- CH∆Ø∆†NG TR√åNH CH√çNH ---
ten_file = "dulieu.csv"
tao_file_csv(ten_file)
doc_va_tinh_tong(ten_file)
```

```
#C√¢u 13
import xml.etree.ElementTree as ET
from collections import Counter
import os

# --- T·∫°o file XML nh√≥m thi·∫øt b·ªã ---
def tao_file_nhom_thiet_bi(file_nhom):
    xml_nhom_data = """<?xml version="1.0" encoding="UTF-8"?>
<nhomthietbis>
    <nhom>
        <ma>n1</ma>
        <ten>ƒêi·ªán tho·∫°i</ten>
    </nhom>
    <nhom>
        <ma>n2</ma>
        <ten>M√°y t√≠nh b·∫£ng</ten>
    </nhom>
    <nhom>
        <ma>n3</ma>
        <ten>Laptop</ten>
    </nhom>
</nhomthietbis>
"""
    with open(file_nhom, "w", encoding="utf-8") as f:
        f.write(xml_nhom_data)
    print(f"‚úÖ ƒê√£ t·∫°o file {file_nhom} th√†nh c√¥ng!")

# --- T·∫°o file XML thi·∫øt b·ªã ---
def tao_file_thiet_bi(file_tb):
    xml_tb_data = """<?xml version="1.0" encoding="UTF-8"?>
<thietbis>
    <thietbi manhom="n1">
        <ma>dt01</ma>
        <ten>iPhone 13</ten>
    </thietbi>
    <thietbi manhom="n1">
        <ma>dt02</ma>
        <ten>Samsung S22</ten>
    </thietbi>
    <thietbi manhom="n2">
        <ma>mtb01</ma>
        <ten>iPad Air</ten>
    </thietbi>
    <thietbi manhom="n3">
        <ma>lt01</ma>
        <ten>Macbook Pro</ten>
    </thietbi>
    <thietbi manhom="n3">
        <ma>lt02</ma>
        <ten>Dell XPS 15</ten>
    </thietbi>
</thietbis>
"""
    with open(file_tb, "w", encoding="utf-8") as f:
        f.write(xml_tb_data)
    print(f"‚úÖ ƒê√£ t·∫°o file {file_tb} th√†nh c√¥ng!")


# --- ƒê·ªçc file XML nh√≥m thi·∫øt b·ªã ---
def doc_nhom_thiet_bi(file_nhom):
    tree = ET.parse(file_nhom)
    root = tree.getroot()
    nhoms = {}
    for nhom in root.findall('nhom'):
        ma = nhom.find('ma').text
        ten = nhom.find('ten').text
        nhoms[ma] = ten
    return nhoms

# --- ƒê·ªçc file XML thi·∫øt b·ªã ---
def doc_thiet_bi(file_tb):
    tree = ET.parse(file_tb)
    root = tree.getroot()
    ds_tb = []
    for tb in root.findall('thietbi'):
        manhom = tb.attrib['manhom']
        ma = tb.find('ma').text
        ten = tb.find('ten').text
        ds_tb.append({'manhom': manhom, 'ma': ma, 'ten': ten})
    return ds_tb

# --- Hi·ªÉn th·ªã danh s√°ch nh√≥m ---
def hien_thi_nhom(nhoms):
    print("== Danh s√°ch nh√≥m thi·∫øt b·ªã ==")
    for ma, ten in nhoms.items():
        print(f"{ma} - {ten}")
    print()

# --- Hi·ªÉn th·ªã to√†n b·ªô thi·∫øt b·ªã ---
def hien_thi_thiet_bi(ds_tb):
    print("== Danh s√°ch thi·∫øt b·ªã ==")
    for tb in ds_tb:
        print(f"M√£: {tb['ma']}, T√™n: {tb['ten']}, Nh√≥m: {tb['manhom']}")
    print()

# --- L·ªçc thi·∫øt b·ªã theo m√£ nh√≥m ---
def loc_theo_nhom(ds_tb, manhom):
    print(f"== Thi·∫øt b·ªã thu·ªôc nh√≥m {manhom} ==")
    found = False
    for tb in ds_tb:
        if tb['manhom'] == manhom:
            print(f"  {tb['ma']} - {tb['ten']}")
            found = True
    if not found:
        print(f"  Kh√¥ng t√¨m th·∫•y thi·∫øt b·ªã n√†o thu·ªôc nh√≥m {manhom}")
    print()


# --- T√¨m nh√≥m c√≥ nhi·ªÅu thi·∫øt b·ªã nh·∫•t ---
def nhom_nhieu_thiet_bi(ds_tb):
    if not ds_tb:
        print("== Kh√¥ng c√≥ d·ªØ li·ªáu thi·∫øt b·ªã ƒë·ªÉ t√¨m nh√≥m nhi·ªÅu nh·∫•t ==")
        return

    dem = Counter([tb['manhom'] for tb in ds_tb])
    # Handle case where Counter is empty
    if not dem:
        print("== Kh√¥ng c√≥ d·ªØ li·ªáu thi·∫øt b·ªã ƒë·ªÉ t√¨m nh√≥m nhi·ªÅu nh·∫•t ==")
        return

    nhom_max = dem.most_common(1)[0]
    print(f"== Nh√≥m c√≥ nhi·ªÅu thi·∫øt b·ªã nh·∫•t ==")
    print(f"Nh√≥m: {nhom_max[0]} - S·ªë l∆∞·ª£ng: {nhom_max[1]}")
    print()

# --- Ch∆∞∆°ng tr√¨nh ch√≠nh ---
def main():
    file_nhom = "nhomthietbi.xml"
    file_tb = "thietbi.xml"

    # T·∫°o file n·∫øu ch∆∞a t·ªìn t·∫°i
    if not os.path.exists(file_nhom):
        tao_file_nhom_thiet_bi(file_nhom)
    if not os.path.exists(file_tb):
        tao_file_thiet_bi(file_tb)

    nhoms = doc_nhom_thiet_bi(file_nhom)
    ds_tb = doc_thiet_bi(file_tb)

    hien_thi_nhom(nhoms)
    hien_thi_thiet_bi(ds_tb)
    loc_theo_nhom(ds_tb, "n1")
    nhom_nhieu_thiet_bi(ds_tb)

if __name__ == "__main__":
    main()
```
